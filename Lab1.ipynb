```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_original = cv2.imread('Lena 2.png')
```

```python
print(image_original)
print(image_original.shape)
```

```python
cv2.imshow("Original", image_original)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image_original)
plt.axis("off")
plt.show()
```

```python
image_new = cv2.cvtColor(image_original,cv2.COLOR_RGB2GRAY)
cv2.imshow("Original", image_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_gray_manual = (0.299*image_original[:,:,0] + 0.587*image_original[:,:,1] + 0.114*image_original[:,:,2]).astype(np.uint8)
print(image_gray_manual)
```

```python
cv2.imshow("Original",image_gray_manual)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_manual = (0.101*image_original[:,:,0] + 0.444*image_original[:,:,1] + 0.613*image_original[:,:,2]).astype(np.uint8)
print(image_manual)
cv2.imshow("Original",image_manual)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_gray = cv2.cvtColor(image_original,cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(image_gray,224,255,cv2.THRESH_BINARY) 
cv2.imshow("Result", binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
