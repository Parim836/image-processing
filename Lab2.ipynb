```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
original = cv2.imread('Lena.png').astype(np.int16)
```

```python
result = original - 60
result = np.clip(result,0,255)
cv2.imshow("result", result.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = original + 60
result = np.clip(result,0,255)
cv2.imshow("result", result.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = original / 2
result = np.clip(result,0,255)
cv2.imshow("result", result.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = original * 3
result = np.clip(result,0,255)
cv2.imshow("result", result.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = 255 - original
result = np.clip(result,0,255)
cv2.imshow("result", result.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
CT = cv2.imread('CT_sample 1.tif')
```

```python
CT = cv2.cvtColor(CT, cv2.COLOR_BGR2GRAY)
cv2.imshow("CT", CT)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist = cv2.calcHist([CT],[0],None,[256],[0,256])
print(hist)
y = hist.ravel()
x = np.arange(256)
```

```python

plt.bar(x,y)
plt.show()
```

```python
stretched = cv2.normalize(CT, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
cv2.imshow("stretched", stretched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist = cv2.calcHist([stretched],[0],None,[256],[0,256])
print(hist)
y = hist.ravel()
x = np.arange(256)
```

```python

plt.bar(x,y)
plt.show()
```

```python
equalization = cv2.equalizeHist(CT[:,:,0])
cv2.imshow("equalization", equalization)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist = cv2.calcHist([equalization],[0],None,[256],[0,256])
print(hist)
y = hist.ravel()
x = np.arange(256)
```

```python

plt.bar(x,y)
plt.show()
```
